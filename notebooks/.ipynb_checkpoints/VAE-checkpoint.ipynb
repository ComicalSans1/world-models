```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

import torch.nn as nn
import torch.nn.functional as F

dataset = datasets.MNIST(root="data", download=True, train=True, transform=transforms.ToTensor())
loader = DataLoader(dataset, batch_size=64, shuffle=True)
# ls = [l for l in loader]
```

```python
plt.figure(figsize=(3,3))
plt.imshow(dataset[0][0].squeeze());
```

```python
n_hidden = 8

class Encoder(nn.Module):
    def __init__(self, il=784):
        super().__init__()
        self.hl = nn.Linear(il, 256)
        self.ol = nn.Linear(256, n_hidden)

    def forward(self, x):
        x = torch.relu(self.hl(x))
        x = self.ol(x)
        return x

class Decoder(nn.Module):
    def __init__(self, ol=784):
        super().__init__()
        self.hl = nn.Linear(n_hidden, 256)
        self.ol = nn.Linear(256, ol)

    def forward(self, x):
        x = torch.relu(self.hl(x))
        x = self.ol(x)
        return x

class AutoEncoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = Encoder()
        self.decoder = Decoder()

    def forward(self, x):
        e = self.encoder(x)
        d = self.decoder(e)
        return e, d

model = AutoEncoder()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.MSELoss()

for epoch in range(10):
    total_loss = 0.0
    for batch_idx, data in enumerate(loader):
        # print(batch_idx, data[0][0].squeeze())
        encoded, decoded = model(data[0][0].squeeze().view(-1))
        
        # Compute the loss and perform backpropagation
        # print(decoded.size(), data[0][0].squeeze().view(-1).size())
        loss = criterion(decoded, data[0][0].squeeze().view(-1))
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Update the running loss
        total_loss += loss.item() * data[0][0].squeeze().view(-1).size(0)

    # Print the epoch loss
    epoch_loss = total_loss / len(loader.dataset)
    print(
        "Epoch {}/{}: loss={:.4f}".format(epoch + 1, 10, epoch_loss)
    )
```

```python
# plt.imshow(data[0][0].squeeze());
img = data[0][13].squeeze()
encoded, decoded = model(img.view(-1))

fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.imshow(img)
ax2.imshow(torch.chunk(decoded.detach(), 28));
```

Encoder: Take input and convert to mu and logstd  \
Decoder: Take latent input and reconstruct image  \
VAE: Take output of Encoder, do $z = \mu + \sigma \cdot \epsilon$, pass to Decoder, get output

```python
class Encoder(nn.Module):
    def __init__(self, il=784, n_hidden=8):
        super().__init__()
        self.n_hidden = n_hidden
        self.l1 = nn.Linear(il, 256)
        self.l2 = nn.Linear(256, 128)

        self.mu = nn.Linear(128, n_hidden)
        self.logstd = nn.Linear(128, n_hidden)
        
    def forward(self, x):
        x = torch.relu(self.l1(x))
        x = torch.relu(self.l2(x))

        mu = self.mu(x)
        logstd = self.logstd(x)
        return mu, logstd

class Decoder(nn.Module):
    def __init__(self, il=784, n_hidden=8):
        super().__init__()
        self.l1 = nn.Linear(n_hidden, 128)
        self.l2 = nn.Linear(128, 256)
        self.l3 = nn.Linear(256, il)
        
    def forward(self, x):
        x = torch.relu(self.l1(x))
        x = torch.relu(self.l2(x))
        x = self.l3(x)
        
        return x

class VAE(nn.Module):
    def __init__(self):
        super().__init__()
        self.Encoder = Encoder(784, n_hidden = 100)
        self.Decoder = Decoder(784, n_hidden = 100)

    def forward(self, x):
        mu, logstd = self.Encoder(x)
        
        std = logstd.exp()
        epsilon = torch.randn_like(std)
        z = mu + epsilon * std

        decoded = self.Decoder(z)
        return mu, logstd, decoded
```

```python
def loss_fn(data, reconst, mu, logstd):
    reconst_loss = reconst_loss = nn.functional.binary_cross_entropy_with_logits(reconst, data, reduction='sum')
    KL_Divergence = -0.5 * torch.sum(2 * logstd - mu ** 2 - torch.exp(2 * logstd) + 1)

    return reconst_loss + KL_Divergence
    
binary_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Lambda(lambda x: (x > 0.65).float())
])
    
dataset = datasets.MNIST(root="data", download=True, train=True, transform=binary_transform)
loader = DataLoader(dataset, batch_size=64, shuffle=True)

model = VAE()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

for epoch in range(10):
    total_loss = 0.0
    for batch_idx, data in enumerate(loader):
        data_i = data[0].view(data[0].size(0), -1)
        optimizer.zero_grad()
        
        mu, logstd, decoded = model(data_i)
        loss = loss_fn(data_i, decoded, mu, logstd)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    epoch_loss = total_loss / len(loader.dataset)
    print(
        "Epoch {}/{}: loss={:.4f}".format(epoch + 1, 10, epoch_loss)
    )
```
